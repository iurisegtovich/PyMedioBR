```python
import numpy as np
from matplotlib import pyplot as plt
% matplotlib notebook
```

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
z=np.cos(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.plot(x, z)
plt.show()

#pergunta: qual Ã© qual!?
```

```python
x = np.arange(0, 10, 0.2)
y = x-1 #                                        reta
z=(x-1)*(x-9) #                                        parÃ¡bola
w=(x-1)*(x-9)*(x-5) #                                        cÃºbica
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.plot(x, z)
ax.plot(x, w)
plt.show()
```

```python
x = np.arange(-10, 10, 0.11)
y1 = x #                                        
y2=x**2
y3=x**0
y4=x**-1
y5=x**-2
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y1)
ax.plot(x, y2)
ax.plot(x, y3)
ax.plot(x, y4)
ax.plot(x, y5)
ax.set_ylim(-10,10)
plt.show()
```

```python
x1 = np.arange(-10, 10, 0.11)
y1 = np.exp(x1) #              

z1= 0*x1
z2= 1e9*x1

x2 = np.arange(0.00001, 10, 0.11)
y2 = np.log(x2)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x1, y1)
ax.plot(x2, y2)

ax.plot(x1, z1)
ax.plot(x1, z2)


ax.set_xlim(-10,10)

ax.set_ylim(-10,10)
plt.show()
```
